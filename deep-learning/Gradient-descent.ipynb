```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python
# y = 2 * x1 + (x2^2) / 2 
alpha=0.001
max_count=100000
x_train=[(1.0,3.0),(2.0,4.0),(3.0,5.0),(5.0,9.0),(8.0,10.0)]
y_train=[6.6,12.2,18.8,50.5,66.3]
error_end=0.5
```

```python
xs=[0.0]*5
ys=[0.0]*5
for i in range(0,5):
    xs[i] =x_train[i][0]
    ys[i] =x_train[i][1]
zs =y_train
fig = plt.figure(2)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c='b', marker='^')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
#plt.show
```

```python
theta0 = 0.5 
theta1 = 0.1
data_len = len(y_train)
cn=0
is_finish=False
while True:
    for i in range(data_len):
        hx=theta0*x_train[i][0]+theta1*(x_train[i][1]**2)
        det_theta =alpha*(hx-y_train[i])
        theta0=theta0-det_theta*x_train[i][0]
        theta1=theta1-det_theta*x_train[i][1] 
        error = 0.0
        for i in range(data_len):
            hx = theta0 * x_train[i][0] + theta1 * (x_train[i][1]**2)
            error = error + ((y_train[i] - hx)**2)/2
        cn = cn + 1
        is_finish=(error<error_end)or(cn>max_count)
        print(is_finish,cn,theta0,theta1)
        if is_finish:
            break
    if is_finish:
        break
            #print(error)            
print ("theta:%f, %f" %(theta0, theta1))
```
