```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.datasets.samples_generator import make_circles
```

```python
with tf.Session() as sess:
    fig,ax=plt.subplots()
    ax.plot(tf.random_normal([100]).eval(),tf.random_normal([100]).eval(),"o")
    ax.set_title("sample random plot for TensorFlow")
    plt.show()
   # plt.savefig("result.png")
```

```python
centers=[(-2,-2),(-2,1.5),(1.5,-2),(2,1.5)]
data,features=make_blobs(n_samples=200,centers=centers,n_features=2,cluster_std=0.8,shuffle=False,random_state=42)
with tf.Session() as sess:
    fig,ax=plt.subplots()
    ax.scatter(np.asarray(data).transpose()[0],np.asarray(data).transpose()[1],marker='o',s=250)
    plt.plot()
```

```python
data,features=make_blobs(n_samples=100,n_features=2,centers=3,cluster_std=1.0,center_box=(-10.0,10.0),shuffle=False,random_state=42)
with tf.Session() as sess:
    fig,ax=plt.subplots()
    ax.scatter(np.asarray(data).transpose()[0],np.asarray(data).transpose()[1],marker='o',s=250)
    plt.plot()
```

```python
data,features=make_circles(n_samples=100,shuffle=False,noise=None,random_state=None,factor=0.8)
with tf.Session() as sess:
    fig,ax=plt.subplots()
    ax.scatter(np.asarray(data).transpose()[0],np.asarray(data).transpose()[1],marker='o',s=250)
    plt.plot()
```
